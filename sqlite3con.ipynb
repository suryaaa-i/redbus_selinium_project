```python
import pandas as pd

df_buses_1 = pd.read_csv("df_A_buses.csv")
df_buses_2 = pd.read_csv("df_Bi_buses.csv")
df_buses_3 = pd.read_csv("df_Ch_buses.csv")
df_buses_4 = pd.read_csv("df_G_buses.csv")
df_buses_5 = pd.read_csv("df_P_buses.csv")
df_buses_6 = pd.read_csv("df_R_buses.csv")
df_buses_7 = pd.read_csv("df_S_buses.csv")
df_buses_8 = pd.read_csv("df_TS_buses.csv")
df_buses_9 = pd.read_csv("df_UP_buses.csv")
df_buses_10 = pd.read_csv("df_WB_buses.csv")

Final_df = pd.concat([df_buses_1, df_buses_2, df_buses_3, df_buses_4,df_buses_5, df_buses_6,
                      df_buses_7, df_buses_8, df_buses_9, df_buses_10], ignore_index=True)


print(Final_df)
```

# Creating a SQLITE3 database

```python
import sqlite3

conn = sqlite3.connect("red_bus_details.db")
my_cursor = conn.cursor()
```

```python
# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS bus_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL,
                  Ratings Float NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')
print("Table Created successfully")
```

```python
import sqlite3

conn = sqlite3.connect("red_bus_details.db")
my_cursor = conn.cursor()

# Create the table if it doesn't exist
create_query = '''CREATE TABLE IF NOT EXISTS bus_details (
    Bus_name TEXT,
    Bus_type TEXT,
    Start_time TEXT,
    End_time TEXT,
    Total_duration TEXT,
    Price REAL,
    Seats_Available INTEGER,
    Ratings REAL,
    Route_link TEXT,
    Route_name TEXT
)'''
my_cursor.execute(create_query)

# Insert data
insert_query = '''INSERT INTO bus_details (
    Bus_name,
    Bus_type,
    Start_time,
    End_time,
    Total_duration,
    Price,
    Seats_Available,
    Ratings,
    Route_link,
    Route_name
) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)'''
data = Final_df.values.tolist()

try:
    my_cursor.executemany(insert_query, data)
    conn.commit()
    print("Values inserted successfully")
except Exception as e:
    print("Error inserting values:", e)

conn.close()
```
